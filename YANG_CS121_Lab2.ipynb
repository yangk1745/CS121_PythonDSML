Overview of All Cases

```python
a=glp1['Serious'].copy() #"Serious" Column
a.to_numpy() #converts to numpy
a1=np.sum(a=='Serious')
a2=np.sum(a=='Non-Serious')
print("Total Cases: ", a1+a2, '//array size: ',np.size(a))
print('Total Serious Cases: ',a1)
print('Total Non-Serious Cases: ',a2)
print('Percent Serious Cases/Total Cases: ',round(a1/(a1+a2)*100,2),'%')

```

```python
#masking
amask=a.copy()
amask[amask=='Serious'] = 1
amask[amask=='Non-Serious'] = 0
print('Total Cases: ', np.size(amask))
print('Total Serious Cases: ', np.sum(amask==1))
print('Total Non-Serious Cases: ', np.sum(amask==0))
```

There were a total of 300,554 safety reports, and 29% of which were significant safety events (serious) such as death and/or hospitalization

```python
b=glp1['api1'].copy() #'API' Column
b.to_numpy()
btot=np.size(b)
print('Total Cases: ', btot)
print('Total Number of cases from dulaglutide: ',np.sum(b==1))
print('Total Number of cases from exenatide: ',np.sum(b==2))
print('Total Number of cases from liraglutide: ',np.sum(b==3))
print('Total Number of cases from semaglutide: ',np.sum(b==4))
print('Total Number of cases from tirzepatide: ',np.sum(b==5))
```

```python
bmask=b.copy()
bmask[bmask==1] = 1
bmask[bmask==2] = 2
bmask[bmask==3] = 3
bmask[bmask==4] = 4
bmask[bmask==5] = 5
print('Total Cases: ', np.size(bmask))
print('Cases from dulaglutide: ', np.sum(bmask==1))
print('Cases from exenatide: ', np.sum(bmask==2))
print('Cases from liraglutide: ', np.sum(bmask==3))
print('Cases from semaglutide: ', np.sum(bmask==4))
print('Cases from tirzepatide: ', np.sum(bmask==5))
```

There were safety reports for all approved GLP-1 agonists, ranging from ~43k events for semaglutide to 79K for exenatide

```python
c=glp1['Sex'].copy() #"Sex" Column
c.to_numpy()
c1=np.sum(c=='Female')
c2=np.sum(c=='Male')
print('Total Cases: ', c1+c2, '//array size: ', np.size(c))
print('Total Female: ',np.sum(c=='Female'))
print('Total Male: ',np.sum(c=='Male'))
print('Percent Female/Total Cases: ',round(c1/(c1+c2)*100,2),'%')
```

```python
cmask=c.copy()
cmask[cmask=='Female'] = 0
cmask[cmask=='Male'] = 1
cmask[cmask=='Not Specified'] = 2
print('Total Cases: ', np.size(cmask))
print('Total Female: ', np.sum(cmask==0))
print('Total Male: ', np.sum(cmask==1))
print('Total Not Specified: ', np.sum(cmask==2))
```

62% of safety reports came from female patients

```python
srsm=amask[(amask==1) & (cmask==1)] #serious and male
srsf=amask[(amask==1) & (cmask==0)] #serious and female
print('Total Serious Cases: ', np.size(amask))
print('Serious and Male: ', 'np.sum',np.sum(srsm), 'np.size',np.size(srsm))
print('Serious and Female: ', 'np.sum',np.sum(srsf), 'np.size',np.size(srsf))
print('Percent Female/Total:', round(np.sum(srsf)/(np.sum(srsf)+np.sum(srsm))*100,2),'%')
```

There were 300,554 serious events (deaths and/or hospitalizations), of which 59% were female. The higher female % is probably just an artifact of women typically being more likely to report AEs and utilize healthcare. (This is disturbing, because it implies that many men are underreporting serious adverse events and are either ignoring it or not reporting it to healthcare professionals.)
